```python
df.drop('BP', axis=1, inplace=True)
```

```python
df
```

```python
column_name = 'Benzene'
mean_value = df[column_name].mean()
df[column_name].fillna(mean_value, inplace=True)
```

```python
columns_to_fill = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'SO2', 'CO', 'Ozone', 'Toluene', 'Eth-Benzene', 'MP-Xylene', 'O-Xylene', 'Temp', 'RH', 'WS', 'WD', 'SR']

for column in columns_to_fill:
    if column in df.columns and pd.api.types.is_numeric_dtype(df[column]):
        mean_value = df[column].mean()
        df[column].fillna(mean_value, inplace=True)
    else:
        print(f"Column '{column}' either doesn't exist in the DataFrame or is not numeric.")

print(df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df['PM2.5'].hist()
plt.show()
```

```python
numerical_columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'SO2', 'CO', 'Ozone', 'Benzene', 'Toluene', 'Eth-Benzene', 'MP-Xylene', 'O-Xylene', 'Temp', 'RH', 'WS', 'WD', 'SR']

num_rows = 4  
num_cols = (len(numerical_columns) - 1) // num_rows + 1

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))

if not isinstance(axes, np.ndarray):
    axes = np.array([axes])

for i, column in enumerate(numerical_columns):
    ax = axes.flatten()[i]
    ax.hist(df[column], bins=20, color='skyblue', edgecolor='black')  
    ax.set_title(column)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')


for i in range(len(numerical_columns), num_rows * num_cols):
    axes.flatten()[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
# List of numerical columns
numerical_columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'SO2', 'CO', 'Ozone', 'Benzene', 'Toluene', 'Eth-Benzene', 'MP-Xylene', 'O-Xylene', 'Temp', 'RH', 'WS', 'WD', 'SR']

# Number of rows and columns for subplots
num_rows = 4  # You can adjust this according to the number of columns you want in each row
num_cols = (len(numerical_columns) - 1) // num_rows + 1

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))

# Flatten axes if necessary
if not isinstance(axes, np.ndarray):
    axes = np.array([axes])

# Plot box plots for each numerical column
for i, column in enumerate(numerical_columns):
    ax = axes.flatten()[i]
    ax.boxplot(df[column])
    ax.set_title(column)
    ax.set_ylabel('Value')

# Hide empty subplots
for i in range(len(numerical_columns), num_rows * num_cols):
    axes.flatten()[i].axis('off')

plt.tight_layout()
plt.show()
```
